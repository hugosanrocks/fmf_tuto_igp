```python
import matplotlib.pyplot as plt

def plot_earthquake_hypocenter(hypocenter, station):
    # Earthquake Hypocenter coordinates (x, z)
    eq_x, eq_z = hypocenter

    # Station coordinates (x, z)
    station_x, station_z = station

    # Plotting the earthquake hypocenter and station
    plt.figure(figsize=(8, 6))
    plt.scatter(eq_x, eq_z, s=200, c='red', marker='*', label='Earthquake Hypocenter')
    plt.scatter(station_x, station_z, s=100, c='blue', label='Station')

    # Drawing lines to connect the hypocenter and station
    plt.plot([eq_x, station_x], [eq_z, station_z], 'k--')

    # Add labels and titles
    plt.xlabel('Distance (km)')
    plt.ylabel('Depth (km)')
    plt.title('Earthquake Hypocenter and Station')
    plt.legend()

    # Show the plot
    plt.grid()
    plt.show()

# Example usage:
hypocenter_coords = (10, -10)  # x=10 km, z=-10 km (Depth is negative)
station_coords = (0, 0)        # x=0 km, z=0 km (Free surface)

plot_earthquake_hypocenter(hypocenter_coords, station_coords)
```
